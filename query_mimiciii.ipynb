### **Query**
#### **Reasons for going to the ER**

```python
diagnoses_icd = pd.read_csv('MIMIC-III/DIAGNOSES_ICD.csv', dtype=str)
d_icd_diagnoses = pd.read_csv('MIMIC-III/D_ICD_DIAGNOSES.csv', dtype=str)
```

```python
print(len(diagnoses_icd),len(d_icd_diagnoses))
```

```python
q = "SELECT ICD9_CODE AS ICD9_CODE, SHORT_TITLE AS DESCRIPTION, SUM(ICD9_CODE) AS TOTAL_EVENTS FROM diagnoses_icd JOIN d_icd_diagnoses USING (ICD9_CODE) GROUP BY ICD9_CODE HAVING SUM(ICD9_CODE) > 1 ORDER BY ICD9_CODE DESC" 
sqldf(q, globals())
```

#### **Patient Demographic Information**

```python
patients = pd.read_csv('MIMIC-III/PATIENTS.csv', dtype=str)
admissions = pd.read_csv('MIMIC-III/ADMISSIONS.csv', dtype=str)
diagnoses_icd = pd.read_csv('MIMIC-III/DIAGNOSES_ICD.csv', dtype=str)
d_icd_diagnoses = pd.read_csv('MIMIC-III/D_ICD_DIAGNOSES.csv', dtype=str)
```

```python
q = "SELECT COUNT(admissions.ETHNICITY) as COUNT, admissions.ETHNICITY FROM patients JOIN admissions USING(subject_id) JOIN diagnoses_icd USING(hadm_id) JOIN d_icd_diagnoses USING(icd9_code) GROUP BY admissions.ETHNICITY ORDER BY COUNT DESC" 
sqldf(q, globals())
```

#### **Top Item-ID Used with Descriptions**

```python
chartevents = pd.read_csv('MIMIC-III/CHARTEVENTS.csv', dtype=str)
d_items = pd.read_csv('MIMIC-III/D_ITEMS.csv', dtype=str)
```

```python
q = "SELECT itemid, COUNT(ietmid) as COUNT, label FROM chartevents INNER JOIN d_items USING(itemid) GROUP BY itemid ORDER BY itemid DESC" 
sqldf(q, globals())
```

#### **Ages of Patients Who Died**

```python
patients = pd.read_csv('MIMIC-III/PATIENTS.csv', dtype=str)
```

```python
q = "SELECT subject_id, strftime('%Y-%m-%d', DOB) AS Date_of_birth, strftime('%Y-%m-%d', DOD) AS Date_of_death, strftime('%Y-%m-%d', DOD)-strftime('%Y-%m-%d', DOB) AS Age FROM patients WHERE DOD IS NOT NULL" 
sqldf(q, globals())
```

#### **Average Number of Drugs Across All Admissions**

```python
prescriptions = pd.read_csv('MIMIC-III/PRESCRIPTIONS.csv', dtype=str)
```

```python
q = "SELECT AVG(a.Average_Count) FROM (SELECT COUNT(drug) as Average_Count FROM prescriptions GROUP BY hadm_id) a" 
sqldf(q, globals())
```

#### **Count of Patients Diagnoses**

```python
diagnoses_icd = pd.read_csv('MIMIC-III/DIAGNOSES_ICD.csv', dtype=str)
d_icd_diagnoses = pd.read_csv('MIMIC-III/D_ICD_DIAGNOSES.csv', dtype=str)
```

```python
q = "SELECT dx.ICD9_CODE, ddx.SHORT_TITLE, COUNT(distinct dx.SUBJECT_ID) as CNT_DX FROM diagnoses_icd dx JOIN d_icd_diagnoses ddx USING(icd9_code) GROUP BY dx.ICD9_CODE, ddx.SHORT_TITLE ORDER BY COUNT(distinct dx.SUBJECT_ID) DESC" 
sqldf(q, globals())
```

#### **Number of Drugs Given to Patients Per Admission**

```python
prescriptions = pd.read_csv('MIMIC-III/PRESCRIPTIONS.csv', dtype=str)
```

#### **Top 10 Most Common Lab Procedures**

```python
d_icd_procedures = pd.read_csv('MIMIC-III/D_ICD_PROCEDURES.csv', dtype=str)
procedures_icd = pd.read_csv('MIMIC-III/PROCEDURES_ICD.csv', dtype=str)
```

```python
q = "SELECT COUNT(procedures_icd.icd9_code) as occurrences, short_title, long_title FROM d_icd_procedures JOIN procedures_icd on d_icd_procedures.ICD9_CODE = procedures_icd.ICD9_CODE GROUP BY procedures_icd.icd9_code ORDER BY occurrences DESC" 
sqldf(q, globals())
```
